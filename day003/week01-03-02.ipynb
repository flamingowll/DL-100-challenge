```python
def initializer(X,y):
    n_x = X.shape[1]
    n_h = 6
    n_y = y.shape[1]

    W1 = np.random.randn(n_h,n_x) * 0.01
    b1 = np.zeros((n_h, 1))
    W2 = np.random.randn(n_y,n_h) * 0.01
    b2 = np.zeros((n_y,1))
    
    return {"size":[n_x,n_h,n_y],
            "params":[W1, b1, W2, b2]}
```

```python
W1, b1, W2, b2 = initializer(X_tr, y_tr)["params"]
```

```python
def ReLU(z):
    relu = np.maximum(0,z)
    return relu

def ReLU_prime(z):
    result = (z >= 0).astype(int)
    return result

def sigmoid(z):
    s = 1.0/(1+np.exp(-z))
    return s
```

```python
ReLU([1,2,3,4,5,6,-5])
```

```python
def fprop(X, parameters):
    
    W1 = parameters[0]
    b1 = parameters[1]
    W2 = parameters[2]
    b2 = parameters[3]
    
    Z1 = np.dot(W1,X.T)+b1
    #print(Z1.shape)
    A1 = ReLU(Z1)

    #A1 = np.tanh(Z1)
    #print(A1.shape)
    Z2 = np.dot(W2,A1)+b2
    #print(Z2.shape)
    A2 = sigmoid(Z2)
    #print(A2.shape)

    assert(A2.shape == (1, X.shape[0]))
    
    
    return [Z1,A1,Z2,A2]
```

```python
Z1,A1,Z2,A2 = fprop(X_tr, [W1, b1, W2, b2])
```

```python
def cost(A2, y):
    m = y.shape[0]

    cost  = (-1/m) * np.sum((np.multiply(y.T , np.log(A2))) + (np.multiply((1-y).T , np.log(1-A2))))
    J = float(np.squeeze(cost))
    
    return J
```

```python
cost(A2, y_tr)
```

```python
def backward_propagation(parameters, fprop, X, y):
    m = X.shape[0]
    
    W1 = parameters[0]
    b1 = parameters[1]
    W2 = parameters[2]
    b2 = parameters[3]

    Z1 = fprop[0]
    A1 = fprop[1]
    Z2 = fprop[2]
    A2 = fprop[3]

    dZ2 = A2 - y.T
    #print(dZ2.shape)
    dW2 = (1/m) * np.dot(dZ2, A1.T)
    db2 = (1/m) * np.sum(dZ2, axis = 1, keepdims=True)
    dA1 = np.dot(W2.T, dZ2)
    #    dZ1 =  dA1 *  (1-np.power(A1,2))

    dZ1 =  dA1 *  (ReLU_prime(Z1))
    dW1 = (1/m) * np.dot(dZ1,X)
    db1 = (1/m) * np.sum(dZ1, axis = 1, keepdims=True)

    grads = {
        "dW1": dW1,
        "db1": db1,
        "dZ1": dZ1,
        "dW2": dW2,
        "db2": db2,
        "dZ2": dZ2,    
    }
    return grads
```

```python
grads = backward_propagation([W1, b1, W2, b2], [Z1,A1,Z2,A2], X_tr, y_tr)
```

```python
def update_parameters(parameters, grad, learning_rate):

    W1 = parameters[0]
    b1 = parameters[1]
    W2 = parameters[2]
    b2 = parameters[3]

    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]

    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2

    parameters = [W1, b1, W2, b2]
    
    return parameters
```

```python
update_parameters([W1, b1, W2, b2],grads,0.1)
```

```python
def nn_model(X, y, learning_rate= 0.1,num_iter = 10001, print_cost= True):
    np.random.seed(1)

    n_x , n_fl , n_y = initializer(X,y)["size"]

    parameters = initializer(X,y)["params"]

    W1 = parameters[0]
    b1 = parameters[1]
    W2 = parameters[2]
    b2 = parameters[3]

    for i in range(num_iter):
        #print(parameters,"****")

        # fp
        Z1,A1,Z2,A2 = fprop(X,parameters)
        
        # cost
        costf = cost(A2, y)
        
        # bp
        grads = backward_propagation(parameters,[Z1,A1,Z2,A2],X,y)
        
        # gd
        parameters = update_parameters(parameters, grads,learning_rate)


        # print cost
        if print_cost and i%1000 == 0:
            print ("Cost after iteration %i: %f" %(i, costf))
            #print(i,"*****",A2,"****")
            
    return parameters
```

```python
for h in [0.001,0.01,1,10,100]:
    print("h = ", h)
    p = nn_model(X_tr,y_tr,h, num_iter = 2001)
    p
```

```python
def predict(parameters, X):
    
    Z1,A1,Z2,A2 = fprop(X,parameters)

    predictions = (A2>0.5).astype(int)

    return predictions
```
